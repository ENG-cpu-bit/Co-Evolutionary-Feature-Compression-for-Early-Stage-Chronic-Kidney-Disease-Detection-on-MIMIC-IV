This notebook visualizes the ablation study results, showing the contribution of each SGPO v2 component.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import warnings
warnings.filterwarnings('ignore')

plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})
plt.style.use('seaborn-v0_8-whitegrid')
print('Ready.')
```

## 1. Load Ablation Results

```python
with open('../results/ablation_results.json') as f:
    data = json.load(f)

variants = data['variants']
df = pd.DataFrame(variants)

display_cols = ['variant', 'n_features', 'auc_mean', 'sensitivity_mean', 'accuracy_mean', 'runtime_seconds']
print(df[display_cols].to_string(index=False))
```

## 2. Component Contribution Analysis

```python
names = [v['variant'] for v in variants]
short_names = [n.replace('(', '\n(') for n in names]
auc_vals = [v['auc_mean'] for v in variants]
sens_vals = [v['sensitivity_mean'] for v in variants]
feat_vals = [v['n_features'] for v in variants]

fig, axes = plt.subplots(1, 3, figsize=(16, 6))

colors = ['#2563EB'] + ['#6B7280'] * (len(names)-1)

# AUC
bars = axes[0].barh(short_names, auc_vals, color=colors, alpha=0.85)
axes[0].set_xlabel('AUC-ROC')
axes[0].set_title('AUC-ROC by Variant', fontweight='bold')
for i, v in enumerate(auc_vals):
    axes[0].text(v + 0.001, i, f'{v:.4f}', va='center', fontsize=9)

# Sensitivity
bars = axes[1].barh(short_names, sens_vals, color=colors, alpha=0.85)
axes[1].set_xlabel('Sensitivity')
axes[1].set_title('Sensitivity by Variant', fontweight='bold')
for i, v in enumerate(sens_vals):
    axes[1].text(v + 0.001, i, f'{v:.4f}', va='center', fontsize=9)

# Features
bars = axes[2].barh(short_names, feat_vals, color=colors, alpha=0.85)
axes[2].set_xlabel('Feature Count')
axes[2].set_title('Features Selected', fontweight='bold')
for i, v in enumerate(feat_vals):
    axes[2].text(v + 0.3, i, str(v), va='center', fontsize=9)

plt.suptitle('Ablation Study: Component Contributions', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 3. Component Impact Summary

```python
full = variants[0]
print('Component Impact (relative to Full SGPO v2):')
print('=' * 60)
for v in variants[1:]:
    auc_diff = v['auc_mean'] - full['auc_mean']
    sens_diff = v['sensitivity_mean'] - full['sensitivity_mean']
    feat_diff = v['n_features'] - full['n_features']
    print(f"\n{v['variant']}:")
    print(f"  AUC change:   {auc_diff:+.4f}")
    print(f"  Sens change:  {sens_diff:+.4f}")
    print(f"  Feat change:  {feat_diff:+d}")
```

## 4. Findings

The ablation study demonstrates that each component contributes to the final SGPO v2 performance:

1. **SFOA (Feature Selection)** is the primary contributor to feature reduction
2. **DOA (HP Tuning)** compensates for performance loss from feature reduction
3. **Fungal Growth (FGO)** prevents stagnation and enables deeper feature compression

The full framework outperforms all partial variants, justifying the co-evolutionary design.
